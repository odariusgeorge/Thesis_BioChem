```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


labels = ['Age', 'Sperm Volume(ml)', 'pH', 'Sperm Number', 'Motility']
infected = [38.9, 3.5, 7.5, 10, 20]
non_infected = [38.7, 3, 7.45, 15, 20]

x = np.arange(len(labels))  
width = 0.3  

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, infected, width, label='Infected, 24/308')
rects2 = ax.bar(x + width/2, non_infected, width, label='Not Infected, 284/308')

ax.set_ylabel('Mean')
ax.set_title('Comparing infected and not-infected')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
