# Human papillomavirus infection and sperm parameters in patients undergoing IVF treatment. (#Study1)

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


labels = ['Age', 'Sperm Volume(ml)', 'pH', 'Sperm Number', 'Motility']
infected = [38.9, 3.5, 7.5, 10, 20]
non_infected = [38.7, 3, 7.45, 15, 20]

x = np.arange(len(labels))  
width = 0.3  

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, infected, width, label='Infected, 24/308', color = 'yellow')
rects2 = ax.bar(x + width/2, non_infected, width, label='Not Infected, 284/308', color = 'purple')

ax.set_ylabel('Mean')
ax.set_title('Comparing infected and not-infected')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
plt.rcParams["figure.figsize"] = (15,15)

fig.tight_layout()

plt.show()
```

# Age and HPV (#Study4)

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


labels = ['Normal forms', 
          'LNV spermatozoa', 
          'DNA Fragmentation', 
          'Concentration \n 10^6/ml', 
          'Progressive Motility',
          'Leukocytes \n 10^6/ml',
          'Vitality']
lessThan35 = [1.11, 30.4, 15.3, 60.6, 58.9, 0.5, 66.7]
between36And45 = [1.05, 31.4, 16.7, 64.1, 56.3, 0.4, 64.0]
moreThan45 = [0.69, 36.5, 16.7, 61.2, 51.8, 0.3, 60.1]

x = np.arange(len(labels))  
width = 0.3  

fig, ax = plt.subplots()
rects1 = ax.bar(x - width, lessThan35, width, label='age<=35', color = 'yellow')
rects2 = ax.bar(x - width/6, between36And45, width, label='35<age<=45', color = 'purple')
rects3 = ax.bar(x + width/1.25, moreThan45, width, label='age>45', color = 'blue')

ax.set_ylabel('Factors')
ax.set_title('Comparing ageing of patients with HPV')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 3, height),
                    xytext=(0, 5),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
plt.rcParams["figure.figsize"] = (15,15)
fig.tight_layout()
plt.show()
```
